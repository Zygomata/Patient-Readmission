```python
#import all packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Importing the medical data file for cleaning
df = pd.read_csv(r'C:\Users\arjun\OneDrive\Desktop\WGU\D210\Heart Disease\heart_disease_health_indicators_BRFSS2015.csv')
df.info()
```

To clean all the data, I will be converting the categorical variables which are to numerical by ordinal encoding. 

```python
#ordinal encoding using previous PA (D206)
var_cat = ['HeartDiseaseorAttack', 'HighChol', 'Stroke', 'Diabetes','HighBP']
dict_var = {0 :'No', 1:'Yes', 2: 'Yes'}
for i in var_cat:
    df[i] = df[i].replace(dict_var)
#Check if it worked 
df['HighBP']
```

```python
conv= {0: 'Male', 1: 'Female'}
df['Sex'] = df['Sex'].replace(conv)
df['Sex']
```

```python
#import the cleaned medical data set
df_2 = pd.read_csv(r'C:\Users\arjun\OneDrive\Desktop\WGU\D210\medical_clean.csv')
df_2.info()
```

```python
#adding a source column to make it easier to see where data originates from
df['Source']= 'CDC'
df_2['Source'] = 'WGU'
df.info()
df_2.info()
```

```python
#multiply age by 5 to get a more accurate representation of age so it can be compared to the WGU dataset.
df['Age'] = df['Age'] * 5
df['Age'].head()
```

```python
#combining the data into one dataframe
df_fin = pd.concat([df_2, df], ignore_index=True)
df_fin
```

```python
#check if the source column correctly merged
df_fin['Source']
```

```python
#export the final data frame as a CSV file.
df_fin.to_csv(r'C:\Users\arjun\OneDrive\Desktop\WGU\D210\combined_df.csv', index = False)
```
